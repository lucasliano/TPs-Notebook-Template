```python
# m칩dulos genericos
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# m칩dulos propios (ver splane.py en la carpeta)
from splane import pzmap, grpDelay, bodePlot
```

### Un poco de Latex

$ H(S) = \frac{\frac{K . \omega_{a}}{2} }{S + \frac{\omega_{a}.(K+2)}{2}} \cdot \frac{S - \frac{1}{R3C} }{S + \frac{1}{R3C}} $

```python
# Constantes del circuito
R3 = 1e3 #1k
C = 1e-6 #1uf

# Constantes del OPAMP
K = 100e3 #100k
BW = 10e6  #1Mhz

# Polo operacional real
Wa = BW/K

# Defino los par치metros del diagrama
gain = (K*Wa)/2
zeros = [1/(R3*C)]
poles = [-(Wa*(K+2)/2), -1/(R3*C)]

# Genero la funci칩n transferencia en S
sys = signal.ZerosPolesGain(zeros, poles, gain).to_tf();
```

```python
# Diagrama de polos y ceros
pzmap(sys, 1);
```

```python
# Diagrama de Bode
bodePlot(sys, 2);
```
